# ExtratorDeConteudoDaNasa:

```python
import pandas as pd
import json
import requests
from io import BytesIO
from PIL import Image, ImageFont, ImageDraw
import imageio
import os
```

```python
url = requests.get("http://127.0.0.1:8000/linguagens/?format=json")
```

```python
dicionario_de_filmes = json.loads(url.content)
dicionario_de_filmes
```

```python
conteudos = []
for url in dicionario_de_filmes:
    conteudos = url.get('title')
    conteudos = url.get('url')
    
```

```python
diretorio = input('nome do diret√≥rio: ')

os.makedirs(diretorio)

for url in dicionario_de_filmes:
    print(url.get('title'))
    response = requests.get(url.get('url'))
    imagem = Image.open(BytesIO(response.content))
    draw = ImageDraw.Draw(imagem)
    font = ImageFont.truetype("Uchen-Regular.ttf", 80)
    draw.text((20, 25), "Hello World!", font=font)

    titulo_imagem = f"{url.get('title')}.png"

    imageio.imwrite(f"{diretorio}/{titulo_imagem}", imagem)

    
    
```
